## Problem 1

Design a function that takes a grant sector, maximum grant amount, and a list of community grants and returns the community grants for the given sector of at most the given maximum grant amount.

In particular, the **function MUST operate in two steps**. First, it will find the community grants for the given grant sector, then it will find which of those community grants received at most the maximum grant amount.

Remember to follow the HtDF recipe and the helper rules.

**Thought question for pre-tuorial work:** Which helper rule applies because of the requirement to operate in two steps?

```python
@typecheck
def filter_on_sector_and_grant_amount(sector: str, amt: int, locg: List[CommunityGamingGrant]) -> List[CommunityGamingGrant]:
    """
    return a list of grants from locg that are from the given sector and of at most amt
    """
    #return [] #stub
    # template based on function composition
    
    # You aren't required to include a plan in comments, 
    # and your solution can be all in one line. But..
    # we like the way this looks and reads!
    
    
    # Plan:
    # 1) find the community grants in locg from sector, call it sector_grants
    # 2) find the community grants from sector_grants with grant size at most amt, call it large_sector_grants
    # 3) return large_sector_grants
    
    sector_grants = from_sector_only(sector, locg)
    large_sector_grants = smaller_than_only(amt, sector_grants)
    return large_sector_grants
    
    # The alternate one-line version would be:
    #return smaller_than_only(amt, from_sector_only(sector, locg))
    
    # Can you see how that looks like the three lines above,
    # but with each variable name "substitude" for the expression
    # from which we got its value?

@typecheck
def smaller_than_only(amt: int, locg: List[CommunityGamingGrant]) -> List[CommunityGamingGrant]:
    """
    return a list of grants from locg that are of at most amt
    """
    # return [] #stub
    # template from List[CommunityGamingGrant] with additional parameter
    
    # acc stores the result so far
    acc = [] # type: List[CommunityGamingGrant]
    
    for cg in locg:
        if grant_smaller_than(amt, cg):
            acc.append(cg)
    
    return acc

@typecheck
def grant_smaller_than(i: int, cg: CommunityGamingGrant) -> bool:
    """
    return True if cg has an amount smaller than i and False otherwise
    """
    #return False #stub
    # template from CommunityGamingGrant with additional parameter
    return cg.amt < i

@typecheck
def from_sector_only(sector: str, locg: List[CommunityGamingGrant]) -> List[CommunityGamingGrant]:
    """
    return a list of grants from locg that are from sector
    """
    # return [] #stub
    # template from List[CommunityGamingGrant] with additional parameter
    
    # acc stores the result so far
    acc = [] # type: List[CommunityGamingGrant]
    
    for cg in locg:
        # It is perhaps a judgment call whether we need
        # a helper to check if cg is from sector, but in
        # CPSC 103, we ask you to create this helper!
        if from_sector(sector, cg):
            acc.append(cg)
    
    return acc

@typecheck
def from_sector(sector: str, cg: CommunityGamingGrant) -> bool:
    """
    return True if cg is from the given sector, False otherwise
    """
    #return False #stub
    # template from CommunityGamingGrant with additional parameter
    return cg.sec == sector

start_testing()

expect(filter_on_sector_and_grant_amount("Sport", 20000, []), []) 
expect(filter_on_sector_and_grant_amount("Sport", 30000, L1), [BMX])
expect(filter_on_sector_and_grant_amount("Sport", 55000, L1), [BMX, BARRACUDAS])
expect(filter_on_sector_and_grant_amount("Sport", 51901, L1), [BMX, BARRACUDAS])
expect(filter_on_sector_and_grant_amount("Arts and Culture", 5001, L1), [CHINESE_ORCHESTRA])
expect(filter_on_sector_and_grant_amount("Arts and Culture", 5000, L1), [])

expect(smaller_than_only(1, []), [])
expect(smaller_than_only(100000, L1), L1)
expect(smaller_than_only(30000, L1), [BMX, ARTS_COUNCIL, AFRICAN_DESCENT_SOCIETY, ALL_BODIES, CHINESE_ORCHESTRA])
expect(smaller_than_only(4999, L1), [])

expect(grant_smaller_than(3000, CHINESE_ORCHESTRA), False)
expect(grant_smaller_than(5000, CHINESE_ORCHESTRA), False)
expect(grant_smaller_than(10000, CHINESE_ORCHESTRA), True)
expect(grant_smaller_than(29499, BMX), False)
expect(grant_smaller_than(29500, BMX), False)
expect(grant_smaller_than(29501, BMX), True)

expect(from_sector_only("Sport", []), []) 
expect(from_sector_only("Sport", L1), [BMX, BARRACUDAS])
expect(from_sector_only("Environment", L1), [])
expect(from_sector_only("Human and Social Services", L1), [BIA])

expect(from_sector("Arts and Culture", CHINESE_ORCHESTRA), True)
expect(from_sector("Sport", CHINESE_ORCHESTRA), False)
expect(from_sector("Sport", BMX), True)
expect(from_sector("Human and Social Services", BMX), False)

summary()
```
